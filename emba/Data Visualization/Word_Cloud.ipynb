```python
import requests
from bs4 import BeautifulSoup
```

```python
#所要擷取的網站網址
url = 'https://www.ptt.cc/bbs/Olympics_ISG/M.1628431417.A.72F.html'

#取得網站回傳資料
response = requests.get(url)

#將擷取的網站印出
print(response.text)

#所要擷取的網站網址R-18版本
#url = 'https://www.ptt.cc/bbs/Gossiping/M.1659235086.A.4F5.html'

#設定Header與Cookie
#headers = {'cookie': 'over18=1;'}

#發送get 請求到ppt
#response = requests.get(url, headers = headers)

#將擷取的網站印出
#print(response.text)
```

```python
#指定Lxml作為解析器
soup = BeautifulSoup(response.text, 'lxml')

#使用find_all()找推文
pushes = soup.find_all('div', 'push')

#印出推文
print(pushes)
```

```python
#取得推文內容
for push in pushes:
    message = push.find('span', 'f3 push-content').getText()
    print(message)
```

```python
#將推文寫入檔案中
with open('Olympic2020.txt','w', encoding='utf-8') as fp:
    for push in pushes:
        
        #將冒號和空白去除
        message = push.find('span','f3 push-content').getText().replace(':','').strip()
        print(message)
        fp.write(message + "\n")
```

```python
#導入函式庫
%matplotlib inline
import wordcloud
import matplotlib.pyplot as plt
import numpy as np
from janome.tokenizer import Tokenizer
from PIL import Image
import pandas as pd
```

```python
text_ptt = open('Olympic2020.txt','r', encoding='utf-8').read()

#利用wordcloud函式庫將文字資訊畫成文字雲
wc = wordcloud.WordCloud(width=1000, height=600, background_color="white",
                        font_path=r"C:\Windows\Fonts\msjh.ttc")   #字型的檔案路徑

wc.generate(text_ptt) #產生文字雲
plt.imshow(wc) #畫出文字雲
plt.axis("off")
```
